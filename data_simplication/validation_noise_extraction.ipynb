```python
import numpy as np
import pandas as pd

path = r'./input/验证实验40人.xlsx'
result_path = r'./output/处理后验证集40人.csv'
data = pd.read_excel(path)
result = pd.read_csv(result_path)
data
```

```python
result
noise_result_list = result['noise']
noise_result_array = np.array(noise_result_list)
noise_result_array
```

```python
name = ['name']
sound = list(range(1, 82))
sound = [str(num) for num in sound]
whole_columns = name + sound
whole_columns
```

```python
noise = '请对您听到的声音，表达您的吵闹感受，1表示非常吵闹，9表示一点也不吵闹'
processed_data = pd.DataFrame(columns=whole_columns)
processed_data['name'] = data['您的姓名：']
processed_data
```

```python
for i in range(81):
    sound = str(i + 1)
    if i == 0:
        processed_data[sound] = data[noise]
    else:
        processed_data[sound] = data[noise + '.' + str(i)]

processed_data
```

```python
processed_data['mse'] = 0
processed_data
```

```python
def mse(noise_evaluate_list):
    noise_evaluate_array = np.array(noise_evaluate_list)
    # 计算每个的平方差
    square_diff = np.square(noise_evaluate_array - noise_result_array)
    mse = np.sum(square_diff)
    return mse
```

```python
for i in range(len(processed_data)):
    noise_mse = mse(processed_data.iloc[i][1:-1])
    processed_data.loc[i, 'mse'] = noise_mse
```

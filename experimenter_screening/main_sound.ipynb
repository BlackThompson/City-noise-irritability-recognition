```python
import pandas as pd
import numpy as np

data = pd.read_excel(r'./input/原始数据242.xlsx')
data
```

```python
# get the column index
first_col_1 = data.columns.get_loc('01(水声)')
first_col_last = data.columns.get_loc('01(其他)')
second_col_1 = data.columns.get_loc('02(水声)')
# 其实是9，这小学数学太害人了
per_length = first_col_last - first_col_1
interval = second_col_1 - first_col_1
print(per_length)
print(interval)
```

```python
processed_data = data.loc[:, '您的姓名：']
# transfer to dataframe
processed_data = processed_data.to_frame()
data_length = data.shape[1]

index = first_col_1
while index < data_length - 1:
    # 切记是左闭右开
    temp = data.iloc[:, index:(index + per_length + 1)]
    processed_data = processed_data.join(temp)
    index += interval

print(processed_data.shape[1])
processed_data
```

```python
main_sound_proportion = np.zeros((83, 9))
# main_sound_proportion
```

```python
all_num = processed_data.shape[0]
k = 1
for i in range(83):
    for j in range(9):
        # for k in range(1, processed_data.shape[1]):
        heard_num = len(processed_data[processed_data.iloc[:, k] == 1])
        proportion = heard_num / all_num
        main_sound_proportion[i, j] = proportion
        k += 1
```

```python
main_sound_proportion
```

```python
# 寻找主要声音的最低值
baseline = 0.45
# 没有一个声音超过baseline的index记录
wrong = []
# flag = True表示没有超过baseline

for i in range(83):
    flag = True
    for j in range(9):
        if main_sound_proportion[i, j] > baseline:
            flag = False
    if flag:
        wrong.append(i)
wrong
```

```python
main_sound_df = pd.DataFrame(main_sound_proportion)
sound_category = ['1.水声', '2.鸟鸣虫叫', '3.风声', '4.喧闹声', '5.生活声', '6.音乐歌曲声', '7.汽车声', '8.机器声', '9.其他']
# main_sound_df.set_axis(sound_category, axis=1, inplace=True)
main_sound_df.columns = sound_category
voice_num = []
for i in range(83):
    voice_num.append(i + 1)

main_sound_df.index = voice_num
main_sound_df
```

```python
# slice_position.to_csv(r"./output/slice_position_42_features.csv")
main_sound_df.to_csv(r"./output/main_sound_proportion.csv", encoding='utf_8_sig')
```

```python
# 使用不同的存储空间，不会相互影响
main_sound_onehot_df = main_sound_df.copy(deep=True)
main_sound_onehot_df['main sound'] = 0
# 1表示是主要声音
for i in range(83):
    main = ''
    for j in range(9):
        if main_sound_df.iloc[i, j] > baseline:
            main_sound_onehot_df.iloc[i, j] = 1
        else:
            main_sound_onehot_df.iloc[i, j] = 0

        num = j + 1
        flag = main_sound_onehot_df.iloc[i, j]
        if (flag == 1) & (main == ''):
            main = str(num)
        elif (flag == 1) & (main != ''):
            main = main + ',' + str(num)

    main_sound_onehot_df.loc[i + 1, 'main sound'] = main

main_sound_onehot_df
```

```python
main_sound_onehot_df.to_csv(r"./output/main_sound_onehot.csv", encoding='utf_8_sig')
```

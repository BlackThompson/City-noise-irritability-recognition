```python
# 新建一个dataframe来存储validation的特征
path = './input/validation_xls/ba1-M301-1731.xls'
single_data = extract_features(path)
columns = single_data.columns

validation_features = pd.DataFrame(columns=columns)
validation_features
```

```python
root_dir = r'./input/validation_xls'
# 遍历每一个文件
for root, dirs, files in os.walk(root_dir):
    for file in files:
        path = "{0}\\{1}".format(root, file)
        path = path.replace('\\', '/')
        single_data = extract_features(path)
        validation_features = validation_features.append(single_data, ignore_index=True)
```

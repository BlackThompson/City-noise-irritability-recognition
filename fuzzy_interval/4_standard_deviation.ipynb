```python
import numpy as np
import pandas as pd

cluster = pd.read_excel(r'./input/83_noise_cluster.xlsx')
data_83 = pd.read_csv(r'./input/data_83.csv')

class_1 = []
class_2 = []
class_3 = []
class_4 = []
```

```python
for i in range(len(cluster)):
    if cluster['胡锐吵闹度聚类'].iloc[i] == 1:
        class_1.append(cluster['声音类别'].iloc[i])
    elif cluster['胡锐吵闹度聚类'].iloc[i] == 2:
        class_2.append(cluster['声音类别'].iloc[i])
    elif cluster['胡锐吵闹度聚类'].iloc[i] == 3:
        class_3.append(cluster['声音类别'].iloc[i])
    elif cluster['胡锐吵闹度聚类'].iloc[i] == 4:
        class_4.append(cluster['声音类别'].iloc[i])
```

```python
noise_class = {}
for i in range(len(data_83)):
    if data_83['声音类别'].iloc[i] in class_1:
        noise_class[i] = 1
    elif data_83['声音类别'].iloc[i] in class_2:
        noise_class[i] = 2
    elif data_83['声音类别'].iloc[i] in class_3:
        noise_class[i] = 3
    elif data_83['声音类别'].iloc[i] in class_4:
        noise_class[i] = 4
```

```python
noise_class
```

```python
data_83['吵闹度类别'] = noise_class.values()
```

```python
data_83.to_csv(r'./output/data_83.csv', encoding='utf_8_sig')
```

```python
import pandas as pd
import numpy as np
```

```python
slice_position = pd.read_excel(r".\input\Slice_position_44.xls")
all_data = pd.read_excel(r".\input\6747_sound_features.xls")
```

```python
all_data
```

```python
# 读取所有列标
all_columns = list(all_data)
sound_features_44 = pd.DataFrame(columns=all_columns)
```

```python
sound_features_44
```

```python
for i in range(len(slice_position)):
    id = slice_position.loc[i, "6747_id"]
    temp = all_data[all_data["25s文件"] == id]
    sound_features_44 = sound_features_44.append(temp, ignore_index=True)
```

```python
sound_features_44
```

```python
writer = pd.ExcelWriter(r'./output/44_sound_features.xlsx')
sound_features_44.to_excel(writer)
writer.save()
```
